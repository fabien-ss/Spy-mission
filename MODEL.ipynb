```python
df_cleaned = train_data.dropna()
```

```python
df_cleaned = df_cleaned.drop(index = df_cleaned.loc[(df_cleaned['AMPS'] > 1) | (df_cleaned['AMPS'] < 0)].index)
```

```python
df_cleaned = df_cleaned.drop(index = df_cleaned.loc[(df_cleaned['VOLTS'] > 8.7) | (df_cleaned['VOLTS'] < 0.02)].index)
```

```python
df_cleaned = df_cleaned.drop(index = df_cleaned.loc[(df_cleaned['DELTA'] > 1) | (df_cleaned['DELTA'] < -1)].index)
```

```python
df_cleaned = df_cleaned.drop(index = df_cleaned.loc[(df_cleaned['OUTPUT'] > 9.3) & (df_cleaned['OUTPUT'] < -8.4)].index)
```

```python
df_cleaned = df_cleaned.drop(index = df_cleaned.loc[(df_cleaned['GAMMA'] > 2) & (df_cleaned['GAMMA'] < -2)].index)
```

```python
df_cleaned = df_cleaned.drop(index = df_cleaned.loc[(df_cleaned['TEMP'] > 373.15) & (df_cleaned['TEMP'] < -100)].index)
```

```python
df_cleaned
```

```python
d = df_cleaned.copy()#loc[df_cleaned['UNIT'] == 'K']

def kelvin_to_celsius(kelvin):
    return kelvin - 273.15

def convert_to_celsius(row):
    if(row['UNIT'] == 'K'):
        return kelvin_to_celsius(row['TEMP'])
    elif(row['UNIT'] == 'C'):
        return row['TEMP']
    return row['TEMP']
```

```python
d['TEMP'] = d.apply(convert_to_celsius, axis = 1)
```

```python
kelvin_column = d.loc[(d['UNIT'] == 'K')]['TEMP']
somme = kelvin_column.sum()
mean = somme / len(kelvin_column)
ecart_type = kelvin_column.std()
print('moyenne ', mean)
print('ecart-type ', ecart_type)
```
