```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data.csv', header=None)
```

```python
np_array = df.to_numpy()

import matplotlib.pyplot as plt

plt.figure(figsize=(28, 26))
plt.scatter(np_array[:, 0], np_array[:, 1], cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Visualisation des donnes rduites par PCA en 2D')
plt.show()
```

```python
arr = [-0.49951372 -0.50144689 -0.49776421 -0.49951034 -0.0419198 ]

def compareArray(arr1, arr2):
    for i in range(len(arr1)):
        if(arr1 != arr2):
            return False
    return True

print(np_array[0])
print(arr)
```

```python
# Standardisation des donnes
scaler = StandardScaler()
data_scaled = scaler.fit_transform(np_array)

# Application de PCA
pca = PCA(n_components=2) # Slection de 2 composantes principales
data_pca = pca.fit_transform(data_scaled)

print("Composantes principales :", pca.components_)
print("Variance explique :", pca.explained_variance_ratio_)
print("Donnes transformes :", data_pca)
print(pca)
```

```python
pca_std_devs = np.std(data_pca, axis=0)
print("carts-types des composantes principales :", pca_std_devs)
```
